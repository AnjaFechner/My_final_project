### 5.1. Web Scraping one page

```python
# Website: https://www.goodreads.com/list/show/40351.Books_becoming_movies_2013_2015
url1 = "https://www.goodreads.com/list/show/40351.Books_becoming_movies_2013_2015"  
```

```python
response1 = requests.get(url1)
response1.status_code
```

```python
soup1 = BeautifulSoup(response1.content, "html.parser")
```

```python
# find the css-selector for one book title
soup1.select("tr:nth-child(1) > td:nth-child(3) > a:nth-child(1) > span:nth-child(1)")
```

```python
# find the css-selector for all book title
soup1.select("td:nth-child(3) > a:nth-child(1) > span:nth-child(1)")
```

```python
# find the css-selector for the authors
soup1.select("td:nth-child(3) > span:nth-child(4) > div:nth-child(1) > a:nth-child(1) > span:nth-child(1)")[0].text
```

```python
book_lst1 = soup1.select("td:nth-child(3) > a:nth-child(1) > span:nth-child(1)")
author_lst1 = soup1.select("td:nth-child(3) > span:nth-child(4) > div:nth-child(1) > a:nth-child(1) > span:nth-child(1)")
len(book_lst1), len(author_lst1)
```

```python
from tqdm.notebook import tqdm
Book1 = []
# Yes = 1
Book_into_Movie1 = []

len_book1 = len(book_lst1)

for i in tqdm(range(len_book1)):
    Book1.append(soup1.select("td:nth-child(3) > a:nth-child(1) > span:nth-child(1)")[i].text)
    Book_into_Movie1.append("1")
    
    
```

```python
Book1
```

- transform list into DF and save the list as file

```python
book1 = pd.DataFrame({"Name": Book1, "Book_into_Movie": Book_into_Movie1})
```

```python
book1
```

```python
book1.to_excel("Book_into_movie_1.xlsx")
```

### 5.3. Web Scraping multiple pages

```python
url3 = "https://www.goodreads.com/list/show/252.Books_Made_into_Movies?page=1"    
```

```python
response3 = requests.get(url3)
response3.status_code
```

```python
soup3 = BeautifulSoup(response3.content, "html.parser")
soup3
```

```python
# css-selector for book title -> css selector: .tableList > tbody:nth-child(1) > tr:nth-child(4) > td:nth-child(3) > a:nth-child(1) > span:nth-child(1)
soup3.select("td:nth-child(3) > a:nth-child(1) > span:nth-child(1)")[0].text
```

```python
# css selector for authors -> css selector: .tableList > tbody:nth-child(1) > tr:nth-child(1) > td:nth-child(3) > span:nth-child(4) > div:nth-child(1) > a:nth-child(1) > span:nth-child(1)
soup3.select("td:nth-child(3) > span:nth-child(4) > div:nth-child(1) > a:nth-child(1) > span:nth-child(1)")[0].text
```

```python
# On the list are 818 books / 100 books per page -> 9 pages
iterations3 = range(1, 10)
```

```python
pages3 = []
url_lst3 = []

for i in iterations3:
    url3 = f"https://www.goodreads.com/list/show/252.Books_Made_into_Movies?page={i}"
    url_lst3.append(url3)
    
    # download the url
    response3 = requests.get(url3)
    
    # print the status code
    print("Status code: " + str(response3.status_code))
    
    pages3.append(response3)
    
    # respectful nap
    wait_time = uniform(0.3, 2)
    sleep(wait_time)
```

```python
len(pages3)
```

```python
Book3 = []
# book into movie -> yes = 1
Book_into_Movie3 = []
pages_parsed3 = []


# looping through all pages
for i in range(len(pages3)):
    # a soup for ever search result page
    pages_parsed3.append(BeautifulSoup(pages3[i].content, "html.parser"))
 
    # books
    book_html3 = pages_parsed3[i].select("td:nth-child(3)")
    
    # loop through the html content of the movies of the current page (i)
    for j in range(len(book_html3)):
        
        # extract and append titles to title list
        Book3.append(book_html3[j].select("td:nth-child(3) > a:nth-child(1) > span:nth-child(1)")[0].text)
        Book_into_Movie3.append("1")
```

```python
Book3
```

```python
len(Book3), len(Book_into_Movie3)
```

```python
book3 = pd.DataFrame({"Name": Book3, "Book_into_Movie": Book_into_Movie3})
```

```python
book3.to_excel("Book_into_movie_3.xlsx")
```
