### 4.3 Standardize the columns (rename)
- A small inforamtion: Data wrangling is performed within the steps where it is needed.

```python
file1 = file1.rename(columns={'User Rating': 'Rating'})
file2 = file2.rename(columns={'Book_Title': 'Name', "Num_Customers_Rated": "Reviews"})
file3 = file3.rename(columns={'Price(USD)': 'Price'})
#file4 = file4.rename(columns={"author": "Author", "title": "Name","rating":"Rating", "voters_count": "Reviews", "num_pages":"Pages", "book-description":"intro"  })
#file4 = file4.drop(columns=["Unnamed: 0", "author_url", "url", "gender","hometown","books-num","born-at","works-count","publication-date", "popular-shelves","rating-dist","publisher"])
```

```python
# control the len
len(file1), len(file2), len(file3)
```

### 4.4. Concat the files

```python
data=pd.concat([file1,file2,file3],axis=0)
```

```python
# check the results
data.head()
```

```python
# control the len
len(data)
```

### 4.5. Delete not important columns

The columns "Year", "Genre", "Rank" and "Type" and etc are not important.  We join over the column "Name". The 4 columns will delete.

```python
data = data.drop([ "Genre", "Rank", "Type"], axis=1)
```

```python
data = data.rename(columns={"Name": "Book"})
```

```python
data = data.drop_duplicates("Book")
```

```python
# save the file (step betwwen to safe the dataset)
data.to_excel("Amazon_bs_reviews.xlsx")
```

### 4.6. Data Wrangling

```python
data['Name'] = [str(x).lower().strip() for x in data['Book']]
```

```python
data['Name'] = [str(x).replace(":", "") for x in data['Name']]
```

```python
data['Name'] = [str(x).replace("a novel", "") for x in data['Name']]
```

```python
data['Name'] = [str(x).replace("(", "") for x in data['Name']]
```

```python
data['Name'] = [str(x).replace(")", "") for x in data['Name']]
```

```python
data['Name'] = [str(x).replace(",", "") for x in data['Name']]
```

```python
data['Name'] = [str(x).replace("signet classics", "") for x in data['Name']]
```

```python
data['Name'] = [str(x).replace("The Twilight Saga Book 4", "") for x in data['Name']]
```

```python
data.sort_values(by=['Year'], inplace=True, ascending=False)
```

```python
data.sort_values(by=['Name'], inplace=True, ascending=True)
```
